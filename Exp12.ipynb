```python
#second dataset
import pandas as pd

df = pd.read_csv("/content/Expt11.csv")
print(df)
```

```python
# Count students in each grade
#find how many students in each grade

print(df['Grade'].value_counts())
```

```python
# count number of students in each department

print(df['Department'].value_counts())
```

```python
# percentage distribution of grades

print(df['Grade'].value_counts(normalize=True)*100)
```

```python
# Gender vs garde analysis

print(pd.crosstab(df['Gender'], df['Grade']))
```

```python
# department vs gender

print(pd.crosstab(df['Department'],df['Gender']))
```

```python
# department vs grade

print(pd.crosstab(df['Department'],df['Grade']))
```

```python
# department vs grade percentage

print(pd.crosstab(df['Department'],df['Grade'],normalize='index')*100)
```

```python
# Grouping data

print(df.groupby('Department')['Grade'].value_counts())
```

```python
import pandas as pd

data = {
'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10'],

'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],

'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],

'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],

'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']
}

df = pd.DataFrame(data)

print(df)
```

```python
# frequency count
# count how many orders in each category

print(df['Category'].value_counts())
```

```python
# unique values
# display all unique category

print(df['Category'].unique())
```

```python
# cross tabulation
#category vs payment method

print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
# percentage distribution

print(df['Category'].value_counts(normalize=True)*100)
```

```python
# Filtering Data
# orders from electronics category

electronics_orders = df[df['Category']=='Electronics']
print(electronics_orders)
```

```python
# sorting data
sorted_df = df.sort_values(by='Category')
print(sorted_df)
```
